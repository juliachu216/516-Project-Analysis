```python
import glob
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
all_files = glob.glob('19-441C/*.CSV')
df1 = pd.concat((pd.read_csv(f, index_col=0, header=None) for f in all_files), axis=1, ignore_index=True)
df1['diagnosis']=0

all_files = glob.glob('19-444C+/*.CSV')
df2 = pd.concat((pd.read_csv(f, index_col=0, header=None) for f in all_files), axis=1, ignore_index=True)
df2['diagnosis']=1
```

```python
np.array(df1.index).ravel()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

mpl.rcParams['figure.figsize'] = 10, 8
mpl.rcParams['legend.fontsize'] = 10

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.array(df1.index).ravel()
y1 = np.array(df1['diagnosis'])
y2 = np.array(df2['diagnosis'])
z1 = np.array(df1[15])
z2 = np.array(df2[3])

# plot test data
ax.plot(x, y1, z1, label='Deparaffined negative sample 19-441C')
ax.plot(x, y2, z2, label='Deparaffined positive sample 19-444C+')
ax.legend()

# make labels
ax.set_xlabel('Wavenumber (cm-1)')
ax.set_zlabel('Intensity (cps)')
ax.yaxis.set_ticklabels(['','1','','','','','2'])
ax.legend()
ax.view_init(elev=20., azim=-45)

plt.show()
```
